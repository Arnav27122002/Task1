## **Importing libraries** ##

```python

import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
##
```

```python
# Reading data from remote link
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
print("Data imported successfully")
data.head(10)


```

## **plotting our data points on 2-D graph**

```python
# Plotting the distribution of scores
data.plot(x='Hours', y='Scores', style='o', c='g')  
plt.title('Hours studied vs Percentage score')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

***there is a positive linear relation between the number of hours studied and percentage of score.***

## **Extracting Data** ##

```python
X = data.iloc[:, :-1].values 
y = data.iloc[:, 1].values  
```

 splitting this data into training and test sets, We'll do this by using Scikit-Learn's built-in train_test_split() method:

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0) 
```

### **Training the Algorithm**
 

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 

print("Training complete.")
```

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X, y, c='g')
plt.plot(X, line);
plt.show()
```

### **Making Predictions**
Testing sample data

```python
print(X_test) # Testing data - In Hours
y_pred = regressor.predict(X_test) # Predicting the scores
```

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

## **Testing my data** ##

```python

hrs = 9.25
prediction= regressor.predict(np.array([hrs]).reshape(-1,1))
print(f"no of hours={hrs}")
print(f"predicted percentage={prediction[0]}")

```

### **Evaluating the model**



```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```

## **Conclusion** ##
This predictive task using supervised machine learning concept, predicts the score a student would get based on the number of hours of studying. This linear regression model made using libraries such as numpy, pandas, matplotlib, sklearn etc shows near accurate results with a positive correlation between scores and hours as can be seen by the graph. The model successfully predicted a score of 93.69 if a student studies for 9.25hrs, the mean absolute error of the model is 4.18 indicating that the predictions can deviate by a magnitude of 4.18.
